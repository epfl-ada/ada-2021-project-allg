```python
def gender(x):
    
    'Function to convert wikidata id into a gender string value'
    'We consider male and female as male/female and transgender male/female respectly'

    if x == "['Q6581097']" or x =="['Q2449503']": #males and transgender males
        return 'male'
    elif x == "['Q6581072']" or  x ==  "['Q1052281']": #females and transgender females
        return 'female'
    elif pd.isna(x):
        return np.nan
    else:
        return 'other'

df['gen'] = df['gender'].apply(gender)
df['gen'].value_counts(dropna=False)
```

```python
def my_autopct(pct):
    return ('%.2f' % pct) if pct > 20 else ''
def get_new_labels(sizes, labels):
    new_labels = [label if size > 20 else '' for size, label in zip(sizes, labels)]
    return new_labels
def my_level_list(data):
    list = []
    for i in range(len(data)):
        if (data[i]*100/np.sum(data)) > 2 : #2%
            list.append('Label '+str(i+1))
        else:
            list.append('')
    return list
#plot_birth = df.groupby(['year_of_birth']).sum().plot(kind='pie',subplots='gen',labels=my_level_list(df), autopct=my_autopct)#, labels=get_new_labels(sizes, labels))
#plt.tight_layout()
```
